````python
# ============================================================
# 5. Autograd 计算图深入
# ============================================================

try:
    import torch
    import torch.nn as nn
    print(f"PyTorch version: {torch.__version__}")
except ImportError:
    print("PyTorch 未安装，请先安装: pip install torch")
    raise

# ---- 5.1 手动追踪计算图 ----
# 目标: 理解 PyTorch 如何构建和遍历计算图

print("\n" + "=" * 50)
print("5.1 计算图手动追踪")
print("=" * 50)

# 创建叶子节点 (leaf tensors)
x = torch.tensor([2.0, 3.0], requires_grad=True, name='x')
w = torch.tensor([4.0, 5.0], requires_grad=True, name='w')
b = torch.tensor([1.0], requires_grad=True, name='b')

# 构建计算图: y = (x * w).sum() + b
mul_result = x * w          # 逐元素乘法
sum_result = mul_result.sum() # 求和
y = sum_result + b          # 加偏置

print(f"\nx = {x.tolist()}")
print(f"w = {w.tolist()}")
print(f"b = {b.tolist()}")
print(f"x * w = {mul_result.tolist()}")
print(f"(x * w).sum() = {sum_result.item()}")
print(f"y = (x * w).sum() + b = {y.item()}")

# 反向传播
y.backward()

print(f"\n梯度验证:")
print(f"  dy/dx = w = {w.tolist()}  =>  x.grad = {x.grad.tolist()}  ✓" if x.grad is not None else "  x.grad 未计算!")
print(f"  dy/dw = x = {x.tolist()}  =>  w.grad = {w.grad.tolist()}  ✓" if w.grad is not None else "  w.grad 未计算!")
print(f"  dy/db = 1             =>  b.grad = {b.grad.tolist()}  ✓" if b.grad is not None else "  b.grad 未计算!")

# ---- 5.2 retain_graph 实验 ----
print("\n" + "=" * 50)
print("5.2 retain_graph 实验")
print("=" * 50)

a = torch.tensor([2.0], requires_grad=True)
b = torch.tensor([3.0], requires_grad=True)

# 两个输出共享部分计算图
c = a * b       # c = a * b
d = c * a       # d = a^2 * b

# 第一次 backward，保留计算图
c.backward(retain_graph=True)
print(f"第一次 backward (c = a*b):")
print(f"  a.grad = {a.grad.item()} (期望: b=3.0)")
print(f"  b.grad = {b.grad.item()} (期望: a=2.0)")

# 清零梯度
a.grad = None
b.grad = None

# 第二次 backward
d.backward()
print(f"\n第二次 backward (d = a^2*b):")
print(f"  a.grad = {a.grad.item()} (期望: 2*a*b = {2*2*3:.1f})")
print(f"  b.grad = {b.grad.item()} (期望: a^2 = {2**2:.1f})")

# ---- 5.3 detach 的作用 ----
print("\n" + "=" * 50)
print("5.3 detach 实验")
print("=" * 50)

x = torch.tensor([5.0], requires_grad=True)
y = x * 2
z_detached = y.detach()  # 切断计算图
w = z_detached * 3

print(f"y = x * 2 = {y.item()}, requires_grad={y.requires_grad}")
print(f"z_detached = y.detach() = {z_detached.item()}, requires_grad={z_detached.requires_grad}")
print(f"w = z * 3 = {w.item()}, requires_grad={w.requires_grad}")

# w 对 x 的梯度为 0，因为 detach 切断了链路
w.backward()
print(f"\nw.backward() 后:")
print(f"  x.grad = {x.grad} (期望: None 或 0，因为 detach 切断了梯度流)")

# ---- 5.4 自定义 Function ----
print("\n" + "=" * 50)
print("5.4 自定义 autograd.Function")
print("=" * 50)

class MyReLU(torch.autograd.Function):
    """自定义 ReLU 实现"""
    @staticmethod
    def forward(ctx, input):
        # ctx 用于保存反向传播需要的信息
        ctx.save_for_backward(input)
        return input.clamp(min=0)

    @staticmethod
    def backward(ctx, grad_output):
        # 获取保存的 tensor
        input, = ctx.saved_tensors
        # ReLU 的梯度: 输入 > 0 时为 1，否则为 0
        grad_input = grad_output.clone()
        grad_input[input <= 0] = 0
        return grad_input

# 测试自定义 ReLU
x = torch.tensor([-1.0, 0.0, 1.0, 2.0], requires_grad=True)
y = MyReLU.apply(x)  # 注意: 使用 .apply() 调用
print(f"输入: {x.tolist()}")
print(f"ReLU 输出: {y.tolist()}")

loss = y.sum()
loss.backward()
print(f"梯度: {x.grad.tolist()} (期望: [0, 0, 1, 1])")

print("\n✅ Autograd 计算图实验完成！")
````

````python
# ============================================================
# 6. Transformer 注意力可视化
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F

try:
    import matplotlib.pyplot as plt
    import matplotlib
    matplotlib.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
    matplotlib.rcParams['axes.unicode_minus'] = False
    HAS_MPL = True
except ImportError:
    HAS_MPL = False
    print("matplotlib 未安装，将用文本方式展示")

# ---- 6.1 实现 Multi-Head Attention ----
class SimpleMHA(nn.Module):
    """简化版 Multi-Head Attention，便于理解"""
    def __init__(self, d_model, n_heads):
        super().__init__()
        self.d_model = d_model
        self.n_heads = n_heads
        self.d_k = d_model // n_heads

        self.W_q = nn.Linear(d_model, d_model, bias=False)
        self.W_k = nn.Linear(d_model, d_model, bias=False)
        self.W_v = nn.Linear(d_model, d_model, bias=False)
        self.W_o = nn.Linear(d_model, d_model, bias=False)

    def forward(self, x, return_attention=False):
        batch_size, seq_len, _ = x.shape

        # 线性投影
        Q = self.W_q(x).view(batch_size, seq_len, self.n_heads, self.d_k).transpose(1, 2)
        K = self.W_k(x).view(batch_size, seq_len, self.n_heads, self.d_k).transpose(1, 2)
        V = self.W_v(x).view(batch_size, seq_len, self.n_heads, self.d_k).transpose(1, 2)

        # Scaled Dot-Product Attention
        scores = torch.matmul(Q, K.transpose(-2, -1)) / (self.d_k ** 0.5)
        attn_weights = F.softmax(scores, dim=-1)
        attn_output = torch.matmul(attn_weights, V)

        # 拼接多头
        attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, self.d_model)
        output = self.W_o(attn_output)

        if return_attention:
            return output, attn_weights
        return output

# ---- 6.2 创建一个简单的注意力示例 ----
torch.manual_seed(42)
d_model = 64
n_heads = 4
seq_len = 8

# 模拟一个句子的 token embedding
tokens = ["The", "cat", "sat", "on", "the", "mat", "because", "tired"]
x = torch.randn(1, seq_len, d_model)

# 创建 MHA 并计算注意力
mha = SimpleMHA(d_model, n_heads)
output, attn_weights = mha(x, return_attention=True)

print(f"输入形状: {x.shape}")
print(f"输出形状: {output.shape}")
print(f"注意力权重形状: {attn_weights.shape} (batch, heads, seq, seq)")

# ---- 6.3 可视化注意力模式 ----
if HAS_MPL:
    fig, axes = plt.subplots(1, n_heads, figsize=(4 * n_heads, 4))
    fig.suptitle('Multi-Head Attention Patterns', fontsize=14, fontweight='bold')

    for head_idx in range(n_heads):
        ax = axes[head_idx]
        attn = attn_weights[0, head_idx].detach().numpy()
        im = ax.imshow(attn, cmap='Blues', vmin=0, vmax=1)
        ax.set_xticks(range(seq_len))
        ax.set_yticks(range(seq_len))
        ax.set_xticklabels(tokens, rotation=45, fontsize=8)
        ax.set_yticklabels(tokens, fontsize=8)
        ax.set_title(f'Head {head_idx + 1}', fontsize=10)
        ax.set_xlabel('Key')
        ax.set_ylabel('Query')

    plt.colorbar(im, ax=axes, shrink=0.8)
    plt.tight_layout()
    plt.show()

# ---- 6.4 位置编码对注意力的影响 ----
print("\n" + "=" * 50)
print("6.4 位置编码分析")
print("=" * 50)

def sinusoidal_pe(seq_len, d_model):
    """标准正弦位置编码"""
    pe = torch.zeros(seq_len, d_model)
    position = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)
    div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-torch.log(torch.tensor(10000.0)) / d_model))
    pe[:, 0::2] = torch.sin(position * div_term)
    pe[:, 1::2] = torch.cos(position * div_term)
    return pe

pe = sinusoidal_pe(seq_len, d_model)

# 计算位置编码之间的余弦相似度
pe_norm = F.normalize(pe, dim=-1)
pe_sim = torch.matmul(pe_norm, pe_norm.T)

if HAS_MPL:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # 左图: 位置编码值
    im1 = axes[0].imshow(pe.numpy(), cmap='RdBu', aspect='auto')
    axes[0].set_title('Sinusoidal Positional Encoding', fontweight='bold')
    axes[0].set_xlabel('Embedding Dimension')
    axes[0].set_ylabel('Position')
    plt.colorbar(im1, ax=axes[0])

    # 右图: 位置间余弦相似度
    im2 = axes[1].imshow(pe_sim.numpy(), cmap='Blues', vmin=0, vmax=1)
    axes[1].set_title('Position Cosine Similarity', fontweight='bold')
    axes[1].set_xticks(range(seq_len))
    axes[1].set_yticks(range(seq_len))
    axes[1].set_xlabel('Position')
    axes[1].set_ylabel('Position')
    plt.colorbar(im2, ax=axes[1])

    plt.tight_layout()
    plt.show()

print("\n💡 观察:")
print("  1. 不同 head 学到不同的注意力模式 (local vs global)")
print("  2. 正弦位置编码的相似度随距离衰减但不完全消失")
print("  3. RoPE 通过旋转矩阵直接编码相对位置，更灵活")

print("\n✅ Transformer 注意力可视化实验完成！")
````

## 7. CLIP InfoNCE Loss 推导

### 核心思想

CLIP 通过对比学习，将图像和文本映射到同一个嵌入空间。给定一个 batch 的 $(image, text)$ 对，目标是让匹配的对距离近，不匹配的对距离远。

### 数学推导

给定 batch size $N$，图像嵌入 $\{I_1, ..., I_N\}$ 和文本嵌入 $\{T_1, ..., T_N\}$:

**Step 1: 计算相似度矩阵**

$$S_{ij} = \frac{I_i \cdot T_j}{\|I_i\| \cdot \|T_j\|} \cdot \tau$$

其中 $\tau$ 是温度参数 (learnable temperature)。

**Step 2: InfoNCE Loss**

对于第 $i$ 个图像，匹配正确文本 $T_i$ 的概率:

$$p_i = \frac{\exp(S_{ii})}{\sum_{j=1}^{N} \exp(S_{ij})}$$

损失函数 (图像到文本 + 文本到图像的对称版本):

$$\mathcal{L} = -\frac{1}{2N} \sum_{i=1}^{N} \left[ \log p_i^{(I→T)} + \log p_i^{(T→I)} \right]$$

### 关键理解

- **温度 $\tau$ 的作用**: $\tau$ 小 → 分布更尖锐 (更关注难负样本)，$\tau$ 大 → 分布更平滑
- **Batch size 的影响**: 更大的 batch = 更多负样本 = 更好的对比学习效果
- **与 Cross-Entropy 的关系**: InfoNCE 本质上是一个 $N$ 分类的 Cross-Entropy，其中正样本是 "第 $i$ 个类别"

````python
# ============================================================
# 7. CLIP Loss 代码实现与验证
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F

try:
    import matplotlib.pyplot as plt
    import matplotlib
    matplotlib.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
    matplotlib.rcParams['axes.unicode_minus'] = False
    HAS_MPL = True
except ImportError:
    HAS_MPL = False

# ---- 7.1 实现 CLIP Loss ----
class CLIPLoss(nn.Module):
    """CLIP 对比学习损失函数"""
    def __init__(self, init_temperature=0.07):
        super().__init__()
        # 温度参数是可学习的 (log scale)
        self.log_temperature = nn.Parameter(torch.log(torch.tensor(init_temperature)))

    @property
    def temperature(self):
        return torch.exp(self.log_temperature)

    def forward(self, image_embeds, text_embeds):
        """
        Args:
            image_embeds: [N, D] 图像嵌入
            text_embeds:  [N, D] 文本嵌入
        """
        # L2 归一化
        image_embeds = F.normalize(image_embeds, dim=-1)
        text_embeds = F.normalize(text_embeds, dim=-1)

        # 计算相似度矩阵 [N, N]
        logits = torch.matmul(image_embeds, text_embeds.T) * self.temperature

        # 对角线是正样本
        labels = torch.arange(logits.shape[0], device=logits.device)

        # 对称损失: 图像→文本 + 文本→图像
        loss_i2t = F.cross_entropy(logits, labels)
        loss_t2i = F.cross_entropy(logits.T, labels)

        return (loss_i2t + loss_t2i) / 2

# ---- 7.2 测试 CLIP Loss ----
torch.manual_seed(42)
batch_size = 8
embed_dim = 128

# 模拟 CLIP 的图像和文本嵌入
image_embeds = torch.randn(batch_size, embed_dim, requires_grad=True)
text_embeds = torch.randn(batch_size, embed_dim, requires_grad=True)

clip_loss_fn = CLIPLoss(init_temperature=0.07)
loss = clip_loss_fn(image_embeds, text_embeds)

print(f"Batch size: {batch_size}")
print(f"Embedding dim: {embed_dim}")
print(f"Temperature: {clip_loss_fn.temperature.item():.4f}")
print(f"CLIP Loss: {loss.item():.4f}")
print(f"理论最小值 (完美匹配): {-torch.log(torch.tensor(1.0/batch_size)).item():.4f} = -log(1/N)")

# ---- 7.3 温度参数的影响 ----
print("\n" + "=" * 50)
print("7.3 温度参数分析")
print("=" * 50)

temperatures = [0.01, 0.05, 0.07, 0.1, 0.5, 1.0]
losses = []

for temp in temperatures:
    # 使用固定温度，不更新参数
    image_norm = F.normalize(image_embeds.detach(), dim=-1)
    text_norm = F.normalize(text_embeds.detach(), dim=-1)
    logits = torch.matmul(image_norm, text_norm.T) * temp
    labels = torch.arange(batch_size)
    loss_val = (F.cross_entropy(logits, labels) + F.cross_entropy(logits.T, labels)) / 2
    losses.append(loss_val.item())
    print(f"  τ={temp:.2f}  => Loss={loss_val.item():.4f}")

if HAS_MPL:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # 左图: 温度 vs 损失
    axes[0].plot(temperatures, losses, 'o-', color='#c2553a', linewidth=2, markersize=8)
    axes[0].set_xlabel('Temperature (τ)', fontsize=11)
    axes[0].set_ylabel('InfoNCE Loss', fontsize=11)
    axes[0].set_title('Temperature vs CLIP Loss', fontsize=12, fontweight='bold')
    axes[0].axvline(x=0.07, color='#5a6b4a', linestyle='--', alpha=0.7, label='τ=0.07 (default)')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # 右图: 不同温度下的 softmax 分布
    sample_logits = torch.matmul(image_norm, text_norm.T)[0]  # 第一个样本
    for temp, c in [(0.01, '#c2553a'), (0.1, '#5a6b4a'), (1.0, '#2d2a26')]:
        probs = F.softmax(sample_logits * temp, dim=0).numpy()
        axes[1].bar(range(batch_size), probs, alpha=0.6, label=f'τ={temp}', color=c)
    axes[1].set_xlabel('Text Index', fontsize=11)
    axes[1].set_ylabel('Probability', fontsize=11)
    axes[1].set_title('Softmax Distribution at Different τ', fontsize=12, fontweight='bold')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

print("\n💡 关键观察:")
print("  1. τ 越小，分布越尖锐 (对难负样本更敏感)")
print("  2. τ 越大，分布越平滑 (所有样本被同等对待)")
print("  3. CLIP 论文使用 τ=0.07，说明需要较强的区分能力")
print("  4. 温度是可学习的，训练过程中会自动调整")

print("\n✅ CLIP Loss 推导实验完成！")
````

````python
# ============================================================
# 8. GNN 过平滑实验
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F

try:
    import matplotlib.pyplot as plt
    import matplotlib
    matplotlib.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
    matplotlib.rcParams['axes.unicode_minus'] = False
    HAS_MPL = True
except ImportError:
    HAS_MPL = False

# ---- 8.1 构建一个简单的图 ----
def create_graph(n_nodes=20, n_edges=40):
    """创建一个随机无向图"""
    torch.manual_seed(42)
    edges = torch.randint(0, n_nodes, (2, n_edges))
    # 添加自环
    self_loops = torch.arange(n_nodes).unsqueeze(0).repeat(2, 1)
    edges = torch.cat([edges, self_loops], dim=1)
    return edges

def compute_laplacian_norm(edges, n_nodes):
    """计算对称归一化的邻接矩阵 D^(-1/2) * A * D^(-1/2)"""
    # 构建邻接矩阵
    adj = torch.zeros(n_nodes, n_nodes)
    for i in range(edges.shape[1]):
        adj[edges[0, i], edges[1, i]] = 1.0
        adj[edges[1, i], edges[0, i]] = 1.0

    # 计算度矩阵
    deg = adj.sum(dim=1)
    deg_inv_sqrt = torch.where(deg > 0, deg.pow(-0.5), torch.zeros_like(deg))
    D_inv_sqrt = torch.diag(deg_inv_sqrt)

    # 对称归一化: D^(-1/2) * A * D^(-1/2)
    norm_adj = D_inv_sqrt @ adj @ D_inv_sqrt
    return norm_adj

# ---- 8.2 GCN 层 ----
class SimpleGCNLayer(nn.Module):
    """简单的 GCN 层: H' = σ(D^(-1/2) A D^(-1/2) H W)"""
    def __init__(self, in_features, out_features):
        super().__init__()
        self.linear = nn.Linear(in_features, out_features, bias=False)

    def forward(self, x, norm_adj):
        # 邻居聚合 + 线性变换
        return F.relu(norm_adj @ self.linear(x))

# ---- 8.3 过平滑实验 ----
n_nodes = 20
in_features = 16

# 创建图和初始特征
edges = create_graph(n_nodes)
norm_adj = compute_laplacian_norm(edges, n_nodes)
x_init = torch.randn(n_nodes, in_features)

# 逐层传播，观察节点表示的变化
print("=" * 60)
print("8.3 过平滑实验: 随 GCN 层数增加，节点表示的变化")
print("=" * 60)

x = x_init.clone()
cosine_sims = []  # 记录每层后节点间的平均余弦相似度
variances = []    # 记录每层后特征的方差

# 初始状态
x_norm = F.normalize(x, dim=-1)
init_sim = torch.matmul(x_norm, x_norm.T).mean().item()
cosine_sims.append(init_sim)
variances.append(x.var().item())
print(f"\nLayer 0 (初始):  平均余弦相似度={init_sim:.4f}, 特征方差={x.var().item():.4f}")

# 逐层 GCN 传播 (不训练，只观察信息传播)
gcn_layers = nn.ModuleList([
    nn.Linear(in_features, in_features, bias=False) for _ in range(10)
])

for layer_idx in range(1, 11):
    # GCN 操作: 聚合 + 线性变换 + 激活
    x = norm_adj @ gcn_layers[layer_idx - 1](x)
    x = F.relu(x)

    # 度量
    x_norm = F.normalize(x, dim=-1)
    avg_sim = torch.matmul(x_norm, x_norm.T).mean().item()
    var = x.var().item()
    cosine_sims.append(avg_sim)
    variances.append(var)

    print(f"Layer {layer_idx:2d}:  平均余弦相似度={avg_sim:.4f}, 特征方差={var:.4f}")

# 可视化
if HAS_MPL:
    fig, axes = plt.subplots(1, 3, figsize=(16, 5))

    # 左图: 余弦相似度变化
    axes[0].plot(range(11), cosine_sims, 'o-', color='#c2553a', linewidth=2, markersize=6)
    axes[0].set_xlabel('Number of GCN Layers', fontsize=11)
    axes[0].set_ylabel('Average Cosine Similarity', fontsize=11)
    axes[0].set_title('Over-smoothing: Cosine Similarity', fontsize=12, fontweight='bold')
    axes[0].axhline(y=1.0, color='gray', linestyle='--', alpha=0.5)
    axes[0].grid(True, alpha=0.3)

    # 中图: 特征方差变化
    axes[1].plot(range(11), variances, 's-', color='#5a6b4a', linewidth=2, markersize=6)
    axes[1].set_xlabel('Number of GCN Layers', fontsize=11)
    axes[1].set_ylabel('Feature Variance', fontsize=11)
    axes[1].set_title('Over-smoothing: Feature Variance', fontsize=12, fontweight='bold')
    axes[1].grid(True, alpha=0.3)

    # 右图: 各层节点表示的热力图
    # 收集各层的节点表示
    x_vis = x_init.clone()
    representations = [x_vis.detach().numpy()]
    for layer_idx in range(10):
        x_vis = norm_adj @ gcn_layers[layer_idx](x_vis)
        x_vis = F.relu(x_vis)
        representations.append(x_vis.detach().numpy())

    # 展示第1层和第10层的对比
    import numpy as np
    data_early = representations[1]
    data_late = representations[9]

    axes[2].set_visible(False)
    # 创建子图展示
    gs = fig.add_gridspec(1, 2, left=0.68, right=0.98)
    ax1 = fig.add_subplot(gs[0, 0])
    ax2 = fig.add_subplot(gs[0, 1])

    ax1.imshow(data_early, cmap='RdBu', aspect='auto', vmin=-2, vmax=2)
    ax1.set_title('Layer 1', fontsize=10, fontweight='bold')
    ax1.set_xlabel('Feature Dim')
    ax1.set_ylabel('Node')

    im = ax2.imshow(data_late, cmap='RdBu', aspect='auto', vmin=-2, vmax=2)
    ax2.set_title('Layer 9', fontsize=10, fontweight='bold')
    ax2.set_xlabel('Feature Dim')

    plt.colorbar(im, ax=[ax1, ax2], shrink=0.8)
    plt.tight_layout()
    plt.show()

# ---- 8.4 缓解策略 ----
print("\n" + "=" * 50)
print("8.4 过平滑缓解策略")
print("=" * 50)
print("""
常见缓解方案:
1. 残差连接 (Residual): H' = GCN(H) + H  → 保持原始信息
2. JumpingKnowledge: 选择性聚合各层输出
3. DropEdge: 随机删除边，减少信息混合
4. PairNorm: 层间归一化，防止表示坍缩
5. 使用更浅的模型 + 更大的感受野 (如 GraphSAGE 采样)

核心问题: GCN 的邻居聚合是低通滤波器，层数增加 = 多次滤波 = 高频信号丢失
""")

print("\n✅ GNN 过平滑实验完成！")
````

````python
# ============================================================
# 9. Diffusion 采样对比: DDPM vs DDIM
# ============================================================

import torch
import numpy as np

try:
    import matplotlib.pyplot as plt
    import matplotlib
    matplotlib.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
    matplotlib.rcParams['axes.unicode_minus'] = False
    HAS_MPL = True
except ImportError:
    HAS_MPL = False

# ---- 9.1 噪声调度 (Noise Schedule) ----
def cosine_beta_schedule(timesteps, s=0.008):
    """余弦噪声调度 (改进版，避免线性调度在末尾的跳跃)"""
    steps = timesteps + 1
    x = torch.linspace(0, timesteps, steps)
    alphas_cumprod = torch.cos(((x / timesteps) + s) / (1 + s) * torch.pi * 0.5) ** 2
    alphas_cumprod = alphas_cumprod / alphas_cumprod[0]
    betas = 1 - (alphas_cumprod[1:] / alphas_cumprod[:-1])
    return torch.clip(betas, 0.0001, 0.9999)

# 设置参数
T = 1000  # 总时间步
betas = cosine_beta_schedule(T)
alphas = 1.0 - betas
alphas_cumprod = torch.cumprod(alphas, dim=0)
sqrt_alphas_cumprod = torch.sqrt(alphas_cumprod)
sqrt_one_minus_alphas_cumprod = torch.sqrt(1.0 - alphas_cumprod)

print(f"总时间步: {T}")
print(f"α_0 = {alphas[0].item():.6f}, α_T = {alphas[-1].item():.6f}")
print(f"ᾱ_0 = {alphas_cumprod[0].item():.6f}, ᾱ_T = {alphas_cumprod[-1].item():.6f}")

# ---- 9.2 前向加噪过程 ----
def forward_diffusion(x_0, t, noise=None):
    """
    前向过程: q(x_t | x_0) = N(sqrt(ᾱ_t) * x_0, (1-ᾱ_t) * I)
    直接从 x_0 计算 x_t，无需逐步加噪
    """
    if noise is None:
        noise = torch.randn_like(x_0)
    sqrt_alpha = sqrt_alphas_cumprod[t].reshape(-1, 1)
    sqrt_one_minus_alpha = sqrt_one_minus_alphas_cumprod[t].reshape(-1, 1)
    return sqrt_alpha * x_0 + sqrt_one_minus_alpha * noise, noise

# 可视化前向过程
torch.manual_seed(42)
x_0 = torch.randn(1, 2)  # 2D 数据点
timesteps_show = [0, 100, 200, 500, 700, 999]

print("\n前向加噪过程 (x_0 → x_T):")
for t in timesteps_show:
    x_t, _ = forward_diffusion(x_0, t)
    print(f"  t={t:4d}: x_t = {x_t[0].tolist()}, ||x_t|| = {x_t.norm().item():.4f}")

# ---- 9.3 DDPM 采样 ----
def ddpm_sample_step(model_output, t, x_t):
    """
    DDPM 单步采样:
    x_{t-1} = (1/√α_t) * (x_t - (β_t/√(1-ᾱ_t)) * ε_θ(x_t, t)) + σ_t * z
    其中 z ~ N(0, I), σ_t = √β_t
    """
    beta_t = betas[t]
    alpha_t = alphas[t]
    alpha_cumprod_t = alphas_cumprod[t]

    # 预测的均值
    pred_mean = (1.0 / torch.sqrt(alpha_t)) * (
        x_t - (beta_t / torch.sqrt(1.0 - alpha_cumprod_t)) * model_output
    )

    # 添加噪声 (t=0 时不添加)
    if t > 0:
        noise = torch.randn_like(x_t)
        sigma_t = torch.sqrt(beta_t)
        x_prev = pred_mean + sigma_t * noise
    else:
        x_prev = pred_mean

    return x_prev

# ---- 9.4 DDIM 采样 ----
def ddim_sample_step(model_output, t, t_prev, x_t, eta=0.0):
    """
    DDIM 单步采样 (确定性版本, eta=0):
    x_{t-1} = √ᾱ_{t-1} * pred_x0 + √(1-ᾱ_{t-1}-σ_t²) * ε_θ + σ_t * z
    其中 σ_t = η * √((1-ᾱ_{t-1})/(1-ᾱ_t)) * √(1-ᾱ_t/ᾱ_{t-1})
    eta=0 时为完全确定性采样
    """
    alpha_cumprod_t = alphas_cumprod[t]
    alpha_cumprod_t_prev = alphas_cumprod[t_prev] if t_prev >= 0 else torch.tensor(1.0)

    # 预测 x_0
    pred_x0 = (x_t - torch.sqrt(1.0 - alpha_cumprod_t) * model_output) / torch.sqrt(alpha_cumprod_t)

    # 计算方差
    sigma_t = eta * torch.sqrt(
        (1.0 - alpha_cumprod_t_prev) / (1.0 - alpha_cumprod_t) *
        (1.0 - alpha_cumprod_t / alpha_cumprod_t_prev)
    )

    # 方向指向 x_t
    direction = torch.sqrt(1.0 - alpha_cumprod_t_prev - sigma_t ** 2) * model_output

    x_prev = torch.sqrt(alpha_cumprod_t_prev) * pred_x0 + direction

    if eta > 0 and t > 0:
        x_prev += sigma_t * torch.randn_like(x_t)

    return x_prev

# ---- 9.5 对比实验: 完整采样过程 ----
print("\n" + "=" * 50)
print("9.5 DDPM vs DDIM 采样对比")
print("=" * 50)

# 模拟一个 "完美" 的噪声预测模型 (这里用恒等函数简化)
def dummy_model(x_t, t):
    """模拟噪声预测模型 (简化版)"""
    # 真实场景中这是 UNet，这里简化为返回加权噪声
    return x_t / torch.sqrt(1.0 - alphas_cumprod[t] + 1e-8)

# DDPM: 全步采样 (1000步)
x_ddpm = torch.randn(1, 2)
ddpm_trajectory = [x_ddpm.clone()]

for t in reversed(range(T)):
    model_out = dummy_model(x_ddpm, t)
    x_ddpm = ddpm_sample_step(model_out, t, x_ddpm)
    if t % 100 == 0 or t == 0:
        ddpm_trajectory.append(x_ddpm.clone())

# DDIM: 加速采样 (只需 50 步)
ddim_steps = 50
ddim_timesteps = np.linspace(T - 1, 0, ddim_steps, dtype=int).tolist()

x_ddim = torch.randn(1, 2)
ddim_trajectory = [x_ddim.clone()]

for i in range(len(ddim_timesteps) - 1):
    t = ddim_timesteps[i]
    t_prev = ddim_timesteps[i + 1]
    model_out = dummy_model(x_ddim, t)
    x_ddim = ddim_sample_step(model_out, t, t_prev, x_ddim, eta=0.0)
    ddim_trajectory.append(x_ddim.clone())

print(f"DDPM 最终位置: {x_ddpm[0].tolist()}")
print(f"DDIM 最终位置: {x_ddim[0].tolist()}")
print(f"\nDDPM 采样步数: {T}")
print(f"DDIM 采样步数: {ddim_steps} (加速 {T//ddim_steps}x)")

# 可视化轨迹
if HAS_MPL:
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # DDPM 轨迹
    ddpm_x = [t[0, 0].item() for t in ddpm_trajectory]
    ddpm_y = [t[0, 1].item() for t in ddpm_trajectory]
    axes[0].plot(ddpm_x, ddpm_y, 'o-', color='#c2553a', markersize=4, linewidth=1.5, label='DDPM trajectory')
    axes[0].plot(ddpm_x[0], ddpm_y[0], 's', color='#2d2a26', markersize=10, label='Start (noise)')
    axes[0].plot(ddpm_x[-1], ddpm_y[-1], '*', color='#5a6b4a', markersize=15, label='End (data)')
    axes[0].set_title(f'DDPM Sampling ({T} steps)', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('y')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    axes[0].set_aspect('equal')

    # DDIM 轨迹
    ddim_x = [t[0, 0].item() for t in ddim_trajectory]
    ddim_y = [t[0, 1].item() for t in ddim_trajectory]
    axes[1].plot(ddim_x, ddim_y, 'o-', color='#5a6b4a', markersize=6, linewidth=1.5, label='DDIM trajectory')
    axes[1].plot(ddim_x[0], ddim_y[0], 's', color='#2d2a26', markersize=10, label='Start (noise)')
    axes[1].plot(ddim_x[-1], ddim_y[-1], '*', color='#c2553a', markersize=15, label='End (data)')
    axes[1].set_title(f'DDIM Sampling ({ddim_steps} steps, eta=0)', fontsize=12, fontweight='bold')
    axes[1].set_xlabel('x')
    axes[1].set_ylabel('y')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    axes[1].set_aspect('equal')

    plt.tight_layout()
    plt.show()

print("\n💡 关键对比:")
print("  DDPM: 随机采样，每步加噪声，需要完整 T 步")
print("  DDIM: 确定性采样 (eta=0)，可跳步，大幅加速")
print("  DDIM 是 DDPM 的推广 (eta=1 退化为 DDPM)")

print("\n✅ Diffusion 采样对比实验完成！")
````
